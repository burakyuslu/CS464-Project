````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import string
````

# Load Data

````python
# take in the training data
x = pd.read_csv("sign_mnist_train.csv")
x_train = x.iloc[:, 1:]
y_train = x.iloc[:, 0]
````

````python
# take in the test data
test = pd.read_csv("sign_mnist_test.csv")
x_test = test.iloc[:, 1:]
y_test = test.iloc[:, 0]
````

````python
# take in the test data
test = pd.read_csv("sign_mnist_test.csv")
# test.head()
x_test = test.iloc[:, 1:]
y_test = test.iloc[:, 0]
````

## Train Model

````python
clf = DecisionTreeClassifier(max_depth = 2, random_state = 0)
clf.fit(x_train, y_train)
result = clf.predict(x_test)
````

````python
import graphviz

dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=x_train.columns,
    class_names=list(string.ascii_lowercase),
    filled=True,
    rounded=True,
    special_characters=True,
)
graph = graphviz.Source(dot_data)

````

# Train models with different configurations
### Min samples leaf

````python
for i in range(1, 40, 5):
  for min_leaf_samples in range(1, 200, 40):
    clf2 = DecisionTreeClassifier(
      max_depth = i,
      random_state = 0,
      min_samples_leaf=min_leaf_samples, 
      # max_features=leaf_samples
    )
    clf2.fit(x_train, y_train)
    predictions = clf2.predict(x_test)
    print (f'Depth: {i} - min samples leaf samples: {min_leaf_samples}- Accuracy:', sum(predictions == y_test) / len(y_test))
````

### Min samples split

````python
for i in range(1, 40, 5):
  for min_samples_split in range(2, 120, 15):
    clf2 = DecisionTreeClassifier(
      max_depth = i,
      random_state = 0,
      min_samples_split=min_samples_split, 
    )
    clf2.fit(x_train, y_train)
    predictions = clf2.predict(x_test)
    print (f'Depth: {i} - min samples split: {min_samples_split}- Accuracy:', sum(predictions == y_test) / len(y_test))
````

### Max Features

````python
for i in range(1, 40, 5):
  for max_features in range(1, 200, 40):
    clf2 = DecisionTreeClassifier(
      max_depth = i,
      random_state = 0,
      max_features=max_features
    )
    clf2.fit(x_train, y_train)
    predictions = clf2.predict(x_test)
    print (f'Depth: {i} - max features: {max_features}- Accuracy:', sum(predictions == y_test) / len(y_test))
````
