````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import string
````

# Load Data

````python
# take in the data
x = pd.read_csv("sign_mnist_train.csv")
# x.head()

````

````python
x_train = x.iloc[:, 1:]
# x_train.head()
````

````python
y_train = x.iloc[:, 0]
y_train.head()
````

````python
test = pd.read_csv("sign_mnist_test.csv")
test.head()
x_test = test.iloc[:, 1:]
y_test = test.iloc[:, 0]
````

## Train Model

````python
clf = DecisionTreeClassifier(max_depth = 1, random_state = 0)
clf.fit(x_train, y_train)
result = clf.predict(x_test)
````

````python
# accuracy = len(result[result == y_test] > 0) / len(y_test)
print ('comparison', sum(result == y_test))
print ('actual', len(y_test))
print ('accuracy', sum(result == y_test) / len(y_test))
````

````python
import graphviz

dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=x_train.columns,
    class_names=list(string.ascii_lowercase),
    filled=True,
    rounded=True,
    special_characters=True,
)
graph = graphviz.Source(dot_data)

````

````python
for i in range(1, 40, 5):
  for leaf_samples in range(1, 22, 2):
    clf2 = DecisionTreeClassifier(max_depth = i, random_state = 0, min_samples_leaf=leaf_samples)
    clf2.fit(x_train, y_train)
    predictions = clf2.predict(x_test)
    print (f'Depth: {i} - Leaf samples: {leaf_samples}- Accuracy:', sum(predictions == y_test) / len(y_test))
````

````python
from sklearn.ensemble import RandomForestClassifier

for n in range(1, 200, 25):
  clf = RandomForestClassifier(n_estimators = n) 
  
  # Training the model on the training dataset
  # fit function is used to train the model using the training sets as parameters
  clf.fit(x_train, y_train)
  
  # performing predictions on the test dataset
  y_pred = clf.predict(x_test)
  
  # metrics are used to find accuracy or error
  from sklearn import metrics 
  print()
  
  # using metrics module for accuracy calculation
  print(f"n_estimators: {n} ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
````
